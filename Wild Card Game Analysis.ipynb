```python
#fetch the sleep data
files = glob.glob('heartrate_data/*2019-10-01*') #the json files bear the name of the day they were sampled, so just import the ones we want

file_list = []

for filename in files:
    df = pd.read_json(filename)
    file_list.append(df)

heart_df1 = pd.concat(file_list, axis=0, ignore_index=False, sort=False)
heart_df1.head()
```

```python
#split out the json string into columns

json_struct = json.loads(heart_df1.to_json(orient="records"))    
heart_df_flat = pd.io.json.json_normalize(json_struct) #use pd.io.json

json_struct2 = json.loads(heart_df2.to_json(orient="records"))    
heart_df_flat2 = pd.io.json.json_normalize(json_struct2) #use pd.io.json

json_struct3 = json.loads(heart_df3.to_json(orient="records"))    
heart_df_flat3 = pd.io.json.json_normalize(json_struct3) #use pd.io.json

json_struct4 = json.loads(heart_df4.to_json(orient="records"))    
heart_df_flat4 = pd.io.json.json_normalize(json_struct4) #use pd.io.json

json_struct5 = json.loads(heart_df5.to_json(orient="records"))    
heart_df_flat5 = pd.io.json.json_normalize(json_struct5) #use pd.io.json

json_struct6 = json.loads(heart_df6.to_json(orient="records"))    
heart_df_flat6 = pd.io.json.json_normalize(json_struct6) #use pd.io.json

json_struct7 = json.loads(heart_df7.to_json(orient="records"))    
heart_df_flat7 = pd.io.json.json_normalize(json_struct7) #use pd.io.json

json_struct8 = json.loads(heart_df8.to_json(orient="records"))    
heart_df_flat8 = pd.io.json.json_normalize(json_struct8) #use pd.io.json

heart_df_flat7.head()
```

```python
#set the index on the just_date column; this will make it easier for us to plot  
heart_df1['just_date'] = heart_df1['dateTime'].dt.date
heart_df1.set_index('dateTime')

heart_df2['just_date'] = heart_df2['dateTime'].dt.date
heart_df2.set_index('dateTime')

heart_df3['just_date'] = heart_df3['dateTime'].dt.date
heart_df3.set_index('dateTime')

heart_df4['just_date'] = heart_df4['dateTime'].dt.date
heart_df4.set_index('dateTime')

heart_df5['just_date'] = heart_df5['dateTime'].dt.date
heart_df5.set_index('dateTime')

heart_df6['just_date'] = heart_df6['dateTime'].dt.date
heart_df6.set_index('dateTime')

heart_df7['just_date'] = heart_df7['dateTime'].dt.date
heart_df7.set_index('dateTime')

heart_df8['just_date'] = heart_df8['dateTime'].dt.date
heart_df8.set_index('dateTime')
```

```python
#Copy the df's into new df's so that we won't alter the originals when we mask the date range.
#this lets us do additional charting with the original dataset if we want

#we're also only going to bring across every third row; this is just to tidy up the output, especially since we have so many lines


wc_df = heart_df1[heart_df1.index % 3 == 0]  # Selects every third row starting from 0
wc_df2=heart_df2[heart_df2.index % 3 == 0]
wc_df3=heart_df3[heart_df3.index % 3 == 0]
wc_df4=heart_df4[heart_df4.index % 3 == 0]
wc_df5 = heart_df5[heart_df5.index % 3 == 0]   
wc_df6=heart_df6[heart_df6.index % 3 == 0]
wc_df7=heart_df7[heart_df7.index % 3 == 0]
wc_df8=heart_df8[heart_df8.index % 3 == 0]
```

```python
#set the date range -- remember we have to extend a bit due to the timezone handling
start_date = pd.Timestamp('2019-10-01 17:00:00')
end_date = pd.Timestamp('2019-10-02 23:59:59')

#crop out the dates outside our range

wc_df = wc_df[wc_df['dateTime'].between(start_date, end_date)]
wc_df2=wc_df2[wc_df2['dateTime'].between(start_date, end_date)]
wc_df3 = wc_df3[wc_df3['dateTime'].between(start_date, end_date)]
wc_df4 = wc_df4[wc_df4['dateTime'].between(start_date, end_date)]

wc_df5=wc_df5[wc_df5['dateTime'].between(start_date, end_date)]
wc_df6 = wc_df6[wc_df6['dateTime'].between(start_date, end_date)]
wc_df7 = wc_df7[wc_df7['dateTime'].between(start_date, end_date)]
wc_df8 = wc_df8[wc_df8['dateTime'].between(start_date, end_date)]

#check that all df still have size
#if a df has all its rows cropped out, we don't want to try to chart it


print('wc_df shape: ', wc_df.shape)
print('wc_df2 shape ', wc_df2.shape)
print('wc_df3 shape: ', wc_df3.shape)
print('wc_df4 shape ', wc_df4.shape)

print('wc_df5 shape: ', wc_df5.shape)
print('wc_df6 shape ', wc_df6.shape)
print('wc_df7 shape: ', wc_df7.shape)
print('wc_df8 shape ', wc_df8.shape)
```
